## Investigation Overview


> The presentation focuses on certain variables/features related to the borrower and how they impact the loans taken out by borrowers. The insights shared will transition from univarate to bivarate and then multivarate, just like in the analysis of the dataset.These features/variables are : Loan Status, Is_Borrower_Homeowner, Stated Monthly Income,Term and Prosper Risk Score.

## Dataset Overview

> The dataset used for this Project was a Prosper Loan Dataset provided by Udacity.The dataset contains loan information of approximately 113,937 loans, with 81 variables/features (categorical and numerical) which reference either loan information or borrower information.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loan_data = pd.read_csv('loan_data.csv')
```

## What are the Prosper Risk Score of Borrowers?

Most of the borrowers(12,256 borrowers) were given a prosper risk score of 4

```python
plt.figure(figsize = [10, 6]) 

pros_score = loan_data.prosper_score.value_counts(ascending = False).head(10)

pros_score.plot(kind ="barh")
plt.title("Prosper Score Ratings of Borrowers", size = 15)
plt.xlabel("Fequency", size = 15)
plt.ylabel("Prosper Score",size = 15);
```

## Distribution of Loan Status

The chart reveals that the "current" loan status is the highest amongst all the other loan statutes. Which means 
majority of the borrowers are making payments on time.


```python

plt.figure(figsize = [12, 6]) 

status = loan_data.loan_status.value_counts(ascending = False)

status.plot(kind ="barh")
plt.title("The Statuses of Loans given by Prosper Loanss ", size = 15)
plt.xlabel("Frequency", size = 15)
plt.ylabel("Loan Status",size = 15);
```

## Distribution Borrowers who are HomeOwners.

It can be clearly seen from the piechart that 53% of borrowers are homeowners compared to 47% of borrowers who aren't


```python
homeowner_percentage = loan_data.is_borrower_homeowner.value_counts()

plt.figure(figsize=(11,8))
plt.title("Percentages of Borrowers Who are Homeowners",fontsize=20)
homeowner_percentage.plot(kind='pie',fontsize=12, autopct='%1.0f%%',shadow = True)
plt.xlabel(' ', fontsize=12)
plt.ylabel(' ', fontsize=12);
```

## Distribution Stated Montly Income of Borrowers.

The stated monthly incomes of borrowers shows a right skewed distribution,with most monthly income of borrowers falling between 2,500 and 5,000.

```python
plt.figure(figsize = [10, 8])
bin= np.arange(0, loan_data['stated_monthly_income'].max()+600, 600)
plt.title("Distribtution of the Stated Monthly Income of Borrowers ", size = 15)
plt.hist(data= loan_data, x = 'stated_monthly_income', bins= bin)
plt.xlim(0, 20000)
plt.xlabel('Stated Monthly Income');
```

## Distribution Borrowers Loan Terms .

The common loan term preferred by borrowers is 36-months.

```python
plt.figure(figsize = [10, 7]) 

a = loan_data.term.value_counts(ascending = False)
a.plot(kind ="barh")
plt.title("The Common Loan Terms Expressed in Months", size = 15)
plt.xlabel("Frequency", size = 15)
plt.ylabel("Months",size = 15);
```

## The relationship between the loan status of borrowers and Monthly Income

The box plot reveals that median salary or typical salary of borrowers who are in the Final Payment Stage of the Loan are higher than borrowers with other loan statuses.

```python
base_color = sb.color_palette()[0]
plt.figure(figsize = [14, 8])
sb.boxplot(data= loan_data, y = 'loan_status', x='stated_monthly_income', showfliers=False,color=base_color)
plt.title("The Loan Status of Borrowers and their Typical Monthly Income", size = 15)
plt.ylabel('Loan Status',size =14);
plt.xlabel('Median Monthly Income', size = 14);
```

# The Relationhip between Loan term and Homeowners

The Clustered chart shows that for the 3 different types of loan terms, borrowers who are homeowners prefer to take out a 36-month loan term.

```python
plt.figure(figsize = [12, 8])
sb.countplot(data = loan_data, x = 'term', hue = 'is_borrower_homeowner', palette = 'Blues');
plt.title("The Loan Terms offered to HomeOwner Borrowers and Non-HomeOwners   ", size = 15)
plt.ylabel('Number of Borrowers',size =14);
plt.xlabel('Loan Term Expressed in Months', size = 14);
```

## The Effect of stated monthly income on prosper score and homeownership

The chart reveals an upward trend in monthly income as prosper risk score increases for both homeowners and non Homeowners. With Homeowners earning higher montly salaries than non-homeowners.

```python
plt.figure(figsize = [12, 8])
ax = sb.pointplot(data = loan_data, x = 'prosper_score', y = 'stated_monthly_income', hue = 'is_borrower_homeowner',dodge = 0.3)
plt.title("The Relationship between Stated Monthly Income, Prosper Score and Homeownership ", size = 15)
plt.xlabel('Prosper Risk Score')
plt.ylabel('Borrower Monthly Income');
```
